```python
import json
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import random
import time
import urllib


dir_out = 'yahoo_data'

if not os.path.exists(dir_out):
    print("create directory: {}".format(dir_out))
    os.mkdir(dir_out)
query_url = "http://finance.yahoo.com/q/hp?s={}"
table_url = "http://ichart.finance.yahoo.com/table.csv?s={}&d=6&e=24&f=2013&g=d&a=4&b=18&c=2012&ignore=.csv"
symbols = ['FB', 'AAPL', 'GOOG']

dataframes = {}
for symbol in symbols:
    time.sleep(random.randint(1, 3))
    url = table_url.format(symbol)
    print(url)
    data_out = {}
    data_out.setdefault('less than 15', 0)
    data_out.setdefault('more than 15', 0)
    data = urllib.urlopen(url).read()
    
    data = data.split('\n')
    filepath = os.path.join(dir_out, "{}_data.csv".format(symbol))
    with open(filepath, 'w') as f:
        print('saving {}'.format(filepath))
        f.writelines("\n".join(data))
        f.close()
    dataframes[symbol] = pd.DataFrame.from_csv(filepath)
    for d in data:
        try:
            d = d.split(',')
            close = d[6]
            if float(close) >= float(15):
                data_out['more than 15'] += 1
            elif float(close) < float(15):
                data_out['less than 15'] += 1
        except:
            pass
    print json.dumps(data_out)
    
```

```python
def plot_dataframes(dfs):
    fig, ax = plt.subplots()
    for symbol, df in dfs.items():
        df.plot(y='Close', label=symbol, ax=ax)
    ax.legend()
```

```python
plot_dataframes(dataframes)
```
