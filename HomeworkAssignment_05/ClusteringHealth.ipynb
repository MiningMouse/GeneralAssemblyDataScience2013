### Creating Value Mappings

The columns 'Indicator Code' and 'Indicator Name' represent a one to one mapping.  Each row that has the value 'SP.DYN.CBRT.IN' for the column 'Indicator Code', has the value 'Birth rate, crude (per 1,000 people)' for the column 'Indicator Name'.  This also goes for the columns 'Country Code' and 'Country Name'.  It would be convenient to create these mappings as dictionaries and keep them around.

There are probably many ways to accomplish this.  The strategy used here is:

1. Use column selection to reduce the dataframe to only the columns 'Indicator Code' and 'Indicator Name'.
1. Use the Pandas DataFrame method groupby to make a grouping.  This is similar to the SQL GROUP BY statement.  There are no additional columns to use for aggregation, but that's okay, we are just interested in the groupings.
1. Iterate over the groups and use the name in a list comprehension.

```python
[name for name, group in df[['Indicator Code', 'Indicator Name']].groupby(['Indicator Code', 'Indicator Name'])]
```

Now with just a minor modification we will put this in a dictionary comprehension.  (This requires Python 2.7+ or Python 3.)

```python
{key: value for (key, value), group in df[['Indicator Code', 'Indicator Name']].groupby(['Indicator Code', 'Indicator Name'])}
```

This has been implemented in the code in the methods get_indicator_code_to_name_map and get_country_code_to_name_map.

### Pivot Tables to the Rescue

Let's say we wanted to get a table containing crude birth rate for 2010 by country.  One way to go about that would be to use boolean indexing to select only those rows corresponding to crude birth rate, then selecting only those columns we are intersted in.

```python
df[df['Indicator Code']=='SP.DYN.CBRT.IN']
```

```python
df[df['Indicator Code']=='SP.DYN.CBRT.IN'][['Country Code', 'Indicator Code', '2010']].head(20)
```

The problem with this approach is that we will have to do this for every metric we are interested in.  A better approach is to pivot the data.  This takes the original dataframe and creates a new index, constructs new columns based on the values passed as 'columns', and uses the values from the column names passed as 'values' as the cell values.  Here is an examle where we create an index based on the column 'Country Code', new columns based on the values in 'Indicator Code', and populate the values with those found in the column '2010'.

```python
df_2010 = df.pivot(index='Country Code', columns='Indicator Code', values='2010')
df_2010
```

Now it is much simpler to plot two metrics against one another.

```python
df_2010.plot(x='SP.DYN.CBRT.IN', y='SP.DYN.CDRT.IN', style='.', figsize=(10,8))
```

The data includes values for some aggregated regions.  Since we are only intersted in countries, we will drop these rows from our data.

```python
df_2010_countries = df_2010[df_2010.index.isin(get_aggregated_country_code_list())==False]
df_2010_countries
```

## Visually Searching for Clusters

Before we apply any clustering algorithms to the data, it would be good to get an intuitive sense of what data might have some meaningful clusters.  To start, we can generate many charts and visually inspect them for clusters.  Since there are 105 columns of data, there are about 5,000 pairs possible (105*104/2).  To make this more manageable, we will restrict the search to just a few metrics, using 'SH.MED.NUMW.P3' (the number of murses and midwives per 1,000 people) against every other column.

```python
column_x = 'SH.MED.NUMW.P3'
n_min_points = 20
for  column_name in df_2010_countries.columns:
    title = "{} vs. {}".format(column_x, column_name)
    try:
        n_points = min(df_2010_countries.count()[column_x], df_2010_countries.count()[column_name])
        if n_points >= n_min_points:
            ax = df_2010_countries.plot(x=column_x, y=column_name, style='.', title=title)
            ax.set_ylabel(column_name)
            plt.show()
        else:
            print("too few points for {}: {}".format(column_name, n_points))
    except KeyError:
        print "Error creating graph: {}".format(title)
```
