### Creating Value Mappings

The columns 'Indicator Code' and 'Indicator Name' represent a one to one mapping.  Each row that has the value 'SP.DYN.CBRT.IN' for the column 'Indicator Code', has the value 'Birth rate, crude (per 1,000 people)' for the column 'Indicator Name'.  This also goes for the columns 'Country Code' and 'Country Name'.  It would be convenient to create these mappings as dictionaries and keep them around.

There are probably many ways to accomplish this.  The strategy used here is:

1. Use column selection to reduce the dataframe to only the columns 'Indicator Code' and 'Indicator Name'.
1. Use the Pandas DataFrame method groupby to make a grouping.  This is similar to the SQL GROUP BY statement.  There are no additional columns to use for aggregation, but that's okay, we are just interested in the groupings.
1. Iterate over the groups and use the name in a list comprehension.

```python
[name for name, group in df[['Indicator Code', 'Indicator Name']].groupby(['Indicator Code', 'Indicator Name'])]
```

Now with just a minor modification we will put this in a dictionary comprehension.  (This requires Python 2.7+ or Python 3.)

```python
{key: value for (key, value), group in df[['Indicator Code', 'Indicator Name']].groupby(['Indicator Code', 'Indicator Name'])}
```

This has been implemented in the code in the methods get_indicator_code_to_name_map and get_country_code_to_name_map.

```python
indicator_map = get_indicator_code_to_name_map(df)
```

```python
country_code_map = get_country_code_to_name_map(df)
```

### Pivot Tables to the Rescue

Let's say we wanted to get a table containing crude birth rate for 2010 by country.  One way to go about that would be to use boolean indexing to select only those rows corresponding to crude birth rate, then selecting only those columns we are intersted in.

```python
df[df['Indicator Code']=='SP.DYN.CBRT.IN']
```

```python
df[df['Indicator Code']=='SP.DYN.CBRT.IN'][['Country Code', 'Indicator Code', '2010']].head(20)
```

The problem with this approach is that we will have to do this for every metric we are interested in.  A better approach is to pivot the data.  This takes the original dataframe and creates a new index, constructs new columns based on the values passed as 'columns', and uses the values from the column names passed as 'values' as the cell values.  Here is an examle where we create an index based on the column 'Country Code', new columns based on the values in 'Indicator Code', and populate the values with those found in the column '2010'.

```python
df_2010 = df.pivot(index='Country Code', columns='Indicator Code', values='2010')
df_2010
```

Now it is much simpler to plot two metrics against one another.

```python
df_2010.plot(x='SP.DYN.CBRT.IN', y='SP.DYN.CDRT.IN', style='.', figsize=(10,8))
```

The data includes values for some aggregated regions.  Since we are only intersted in countries, we will drop these rows from our data.

```python
df_2010_countries = df_2010[df_2010.index.isin(get_aggregated_country_code_list())==False]
df_2010_countries
```

## Visually Searching for Clusters

Before we apply any clustering algorithms to the data, it would be good to get an intuitive sense of what data might have some meaningful clusters.  

One way to ensure that we have clusters is to look for metrics that have more than one maxima in their distributions.  If we combine two or more such metrics, we are likely to have clusters, even though their meaningfulness is not guaranteed.

First, an example of a distribution for a single metric, then a call to a function which will plot all the distributions for us.

```python
df_2010_countries['SP.DYN.CBRT.IN'].hist()
```

```python
fig = plot_histograms_for_each_column(df_2010_countries, indicator_map=indicator_map)
```

This gives us a sense for which distributions have a single very tight peak and therefore are probably not good candidates to use as one of the dimensions for clustering.  Those distributions with wide peaks or even multiple maxima might be good candidates.


Another approach is to generate many charts, plotting pairs of metrics, and visually inspect them for clusters.  Since there are 105 columns of data, there are about 5,000 pairs possible (105*104/2).  To make this more manageable, we will restrict the search to just a few metrics, using 'SH.MED.PHYS.ZS' (the number of physicians per 1,000 people) against every other column.

```python
fig = plot_scatter_for_each_pair(df_2010_countries, ('SH.MED.PHYS.ZS',), df_2010_countries.columns)
```

## Clustering with k-means

From the above charts we can get some ideas for candidates for clustering.  For this section, we'll consider the number of physicians per 1,000 people (SH.MED.PHYS.ZS) and the death rate per 1,000 people.

```python
ax = df_2010_countries.plot(x='SH.MED.PHYS.ZS', y='SP.DYN.CDRT.IN', style='.')
ax.set_ylabel('SP.DYN.CDRT.IN')
```

The method cluster_kmeans will use the scikit-learn KMeans clustering algorithm to apply labels to each data point, then plot them color coded based on the label.

```python
fig, df_clusters = cluster_kmeans(df_2010_countries[['SH.MED.PHYS.ZS', 'SP.DYN.CDRT.IN']], n_clusters=3, scale_data=False)
```

Intuitively, we don't expect the points in the upper middle of the graph to be in the same group as those in the upper left of the graph.  The problem here is that our dimensions have very different ranges and KMeans is using Euclidean distance, equating a unit in the number of doctors to a unit in the crude death rate.  One way to mitigate this problem is to scale the data.  The StandardScaler class will scale data in each column to have zero mean and unit variance.

```python
fig, df_clusters = cluster_kmeans(df_2010_countries[['SH.MED.PHYS.ZS', 'SP.DYN.CDRT.IN']], n_clusters=3, scale_data=True)
```

To see what countries are in a particular cluster, we can look at the dataframe returned by this call.

Note that the cluster labels will not be consistent from run to run, even if the countries in a cluster together is consistent.

```python
df_clusters[df_clusters['label']==2].head(20)
```

```python
for country_code in df_clusters[df_clusters['label']==2]['Country Code']:
    print(country_code_map[country_code])
```

This cluster is made up mostly of very poor countries in Africa, as well as Serbia and Afghanistan.

What is the best number of clusters for this data?  One measure of this is the silhouette coefficient, which combines within-cluster distances and without-cluster distances.

```python
get_optimal_clusters(df_2010_countries[['SH.MED.PHYS.ZS', 'SP.DYN.CDRT.IN']], scale_data=True)
```

We see an optimal silhouette coefficient at 3 clusters.

Let's try another pair of features: again the number of physicials per 1,000 people, and the out-of-pocket health expenditure.  First we'll calculate the optimal number of clusters, then we'll plot those clusters.

```python
get_optimal_clusters(df_2010_countries[['SH.MED.PHYS.ZS', 'SH.XPD.OOPC.TO.ZS']], scale_data=True)
```

```python
fig, df_clusters = cluster_kmeans(df_2010_countries[['SH.MED.PHYS.ZS', 'SH.XPD.OOPC.TO.ZS']], n_clusters=10, scale_data=True)
```

Without knowing much about the data here, it's hard to assess whether or not these clusters are meaningful.

As one more example, let's look at the birth rate and death rate by country.

```python
get_optimal_clusters(df_2010_countries[['SP.DYN.CBRT.IN', 'SP.DYN.CDRT.IN']], scale_data=True)
```

While the silhouette scores suggest that 2 is the best cluster number, we'll ignroe this and set it to 5.

```python
fig, df_clusters = cluster_kmeans(df_2010_countries[['SP.DYN.CBRT.IN', 'SP.DYN.CDRT.IN']], n_clusters=5, scale_data=True)
```
